```python
df4 = pd.read_excel("drive/MyDrive/Colab Notebooks/datasets/african-development-bank-bank-operations-2016xlsx-1.xlsx", sheet_name= "Data")

df4
```

```python
df4.drop('Unit',axis=1,inplace=True)
df4
```

```python
def reformat_date(date_str):
    if 'Q' in date_str:
        year, quarter = date_str.split('Q')
        month = int(quarter) * 3
        return f"{year}-{month:02d}-01"
    elif 'M' in date_str:
        year, month_str = date_str.split('M')
        month = int(month_str)
        return f"{year}-{month:02d}-01"
    else:
        return f"{date_str}-01-01"
```

```python
df4['Date'] = df4['Date'].astype(str)
df4
```

```python
df4['Formatted Date'] = df4['Date'].apply(reformat_date)
df4
```

```python
df4['FDate'] = df4['Date'].apply(reformat_date)
df4
```

```python
df4.drop('Country', axis=1, inplace=True)
df4
```

```python
df4.sort_values(by = 'CountryName', inplace = True)
df4
```

```python
df4.drop(['Date','Formatted Date'], axis=1, inplace=True)
df4
```

```python
df4.rename(columns={'FDate':'Date','Indicator':'Indicator_id'},inplace=True)
df4
```

```python
# Filter data for multinational entities in 2009
multinational_2009_df = df4[(df4['CountryName'] == 'Multinational') & (df4['Date'].dt.year == 2009)]

# Group by IndicatorName and sum the Value
indicator_values = multinational_2009_df.groupby('IndicatorName')['Value'].sum()

# Get the top 10 indicators
top_10_multinational_indicators = indicator_values.nlargest(10)

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_10_multinational_indicators, labels=top_10_multinational_indicators.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Top 10 Multinational Indicators by Total Operation Value in 2009')
plt.ylabel('') # Hide the default y-label
plt.show()
```

```python
# Filter out 'Total' and 'Sub-Total' from CountryName
filtered_df4 = df4[~df4['CountryName'].isin(['Total', 'Sub-Total'])]

# Group by CountryName and sum the Value across all years
country_values_filtered = filtered_df4.groupby('CountryName')['Value'].sum()

# Get the top 5 countries from the filtered data
top_5_countries_filtered = country_values_filtered.nlargest(5)

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_5_countries_filtered, labels=top_5_countries_filtered.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Top 5 Countries by Total Operation Value ')
plt.ylabel('') # Hide the default y-label
plt.show()
```

```python
# Filter data for the year 2009
operations_2009_df = df4[df4['Date'].dt.year == 2009]

# Filter out 'Total' and 'Sub-Total' from CountryName
filtered_operations_2009_df = operations_2009_df[~operations_2009_df['CountryName'].isin(['Total', 'Sub-Total'])]

# Group by CountryName and sum the Value
country_values_2009_filtered = filtered_operations_2009_df.groupby('CountryName')['Value'].sum()

# Get the top 5 countries
top_5_countries_2009_filtered = country_values_2009_filtered.nlargest(5)

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_5_countries_2009_filtered, labels=top_5_countries_2009_filtered.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Top 5 Countries by Total Operation Value in 2009 ')
plt.ylabel('') # Hide the default y-label
plt.show()
```

```python
# Extract the year from the 'Date' column
df4['Year'] = df4['Date'].dt.year

# Group by Year and sum the Value
yearly_operations = df4.groupby('Year')['Value'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(15, 7))
sns.lineplot(x='Year', y='Value', data=yearly_operations, marker='o')
plt.title('Total Operation Value Over Time (Yearly)')
plt.xlabel('Year')
plt.ylabel('Total Value')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter data for multinational projects
multinational_projects_df = df4[df4['CountryName'] == 'Multinational']

# Group by Year and sum the Value
multinational_time_series = multinational_projects_df.groupby('Year')['Value'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(15, 7))
sns.lineplot(x='Year', y='Value', data=multinational_time_series)
plt.title('Total Operation Value for Multinational Projects Over Time')
plt.xlabel('Year')
plt.ylabel('Total Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter data for the year 2009
operations_2009_df = df4[df4['Date'].dt.year == 2009]

# Group by IndicatorName and sum the Value
indicator_sum_2009 = operations_2009_df.groupby('IndicatorName')['Value'].sum().nlargest(10)

# Create the bar chart
plt.figure(figsize=(15, 8))
sns.barplot(x=indicator_sum_2009.index, y=indicator_sum_2009.values, palette='viridis', hue=indicator_sum_2009.index, legend=False)
plt.title('Top 10 Indicators by Total Operation Value in 2009')
plt.xlabel('Indicator Name')
plt.ylabel('Total Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Merge df4 with df2 to get region information
df_merged = pd.merge(df4, df2[['Country_name', 'Region']], left_on='CountryName', right_on='Country_name', how='left')

# Group by Region and sum the Value
region_sum = df_merged.groupby('Region')['Value'].sum().sort_values(ascending=False)

# Create the bar chart
plt.figure(figsize=(15, 8))
sns.barplot(x=region_sum.index, y=region_sum.values, palette='viridis')
plt.title('Total Operation Value by Region')
plt.xlabel('Region')
plt.ylabel('Total Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by IndicatorName and sum the Value
indicator_sum_bottom10 = df4.groupby('IndicatorName')['Value'].sum().nsmallest(10)

# Create the bar chart
plt.figure(figsize=(15, 8))
sns.barplot(x=indicator_sum_bottom10.index, y=indicator_sum_bottom10.values, palette='viridis')
plt.title('Bottom 10 Indicators by Total Operation Value')
plt.xlabel('Indicator Name')
plt.ylabel('Total Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
